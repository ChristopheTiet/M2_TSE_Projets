```python
#Creation de list pour stocker les différentes variables

Host=[] #stock l'adresse des visiteurs
Time=[] #stock la date et l'heure des visites
Request=[] #stock les requests GET POST ou AUTRES
Resources=[] #stock l'adresse de la page visitée
HTTP=[] #stock l'info HTTP
ReplyCode=[] #stock le code de réponse
FileSize=[] #stock la taille de la ressource 
Exceptions=[] #stock les données "sales"

#ouverture du fichier et tentative de split dans les listes définis
#Problème lors du split car certaines lignes n'ont pas le même nombre d'éléments(Exceptions).

with open("C:\\Users\\Chris\\Desktop\\Prepa_M2\\Initiation_Python\\data_project.txt","r") as file:
    CountLine=0
    for line in file:
        SplitData=line.split(" ")
        CountLine=CountLine+1
        if len(SplitData)==7:
            Host.append(SplitData[0])
            Time.append(SplitData[1])
            Request.append(SplitData[2])
            Resources.append(SplitData[3])
            HTTP.append(SplitData[4])
            ReplyCode.append(SplitData[5])
            FileSize.append(SplitData[6])
        else: 
            Exceptions.append(SplitData)
        
#Le dataset contient 47 748 observations
print(CountLine)

#On possède 47619 observations "bien" splittées/"propre"
print(len(Host))

#On se retrouve avec 129 observations mal splitées/"sales" à observer de plus près'
print(len(Exceptions))

#J'utiliserai pendant tout le project le terme de données propres ou données sales!
```

```python
#INSTRUCTION 2

#On va essayer de retrouver toutes les codes de réponses uniques à travers la fonction unique()

def unique(list):
    UniqueVar=[]
    for x in list:
        if x not in UniqueVar:
            UniqueVar.append(x)
            
    return UniqueVar

#Extractions des ReplyCode des observations "sales"
    
ExceptionsReplyCode=[]
for element in Exceptions:
    ExceptionsReplyCode.append(element[-2])

#Rajout des ReplyCodes des 129 observations mal splitées avec celles bien splitées.

ReplyCode.extend(ExceptionsReplyCode)

#Création d'une liste UniqueReplyCode avec toutes les variables uniques des codes de réponses.

UniqueReplyCode=unique(ReplyCode)
print(UniqueReplyCode)

#Comptage du nombre d'accès par code retourné pour les observations propres.

for unicode in UniqueReplyCode:
    RequestNumber=ReplyCode.count(unicode)
    print("La Requete {} a eu {} acces".format(unicode,RequestNumber))

```

```python
#INSTRUCTION 3

#Extraction de la variable Time des observations sales

ExceptionsTime=[]
for element in Exceptions:
    ExceptionsTime.append(element[1])

#Rajout de la variable Time des observations sales avec les observations propres

Time.extend(ExceptionsTime)


#Création d'une liste pour y stocker les horaires d'accès selon le jour
Day29=[]
Day30=[]


for element in Time:
    if "[29:" in element:
        Day29.append(element)
    else:
        Day30.append(element)

#Nombre de requête le 29 à partir de 23h

print("Le 29 à 23H, il y a eu {} requêtes dans l'heure.".format(len(Day29)))

#Création d'une liste pour y stocker les heures dans la journée du 30

Day30Hour=[]

for element in Day30:
    Day30Hour.append(element[4:6])

#Création d'une liste pour obtenir une heure unique dans la journée du 30

UniqueHour=unique(Day30Hour)

#Creation d'une liste contenant le nombre de requete chaque heure qui nous servira dans l'instruction 4 

RequestEveryHour=[0,63]

#Comptage du nombre de requête heure par heure

for element in UniqueHour:
    
    count=Day30Hour.count("{}".format(element))
    print("Le 30 à {}H , il y a eu {} requêtes dans l'heure.".format(element,count))
    RequestEveryHour.append(count)
    

print(RequestEveryHour)

#On vérifie qu'on a bien nos 47 748 observations 

print(sum(RequestEveryHour))

    
```

```python
#INSTRUCTION 6

#Creation d'une fonction permettant de compter le nombre de fichier de type x

def file_count(x,l):
    """fonction permettant de compter le nombre de fichier de type x d'une liste l"""

# l'extension du fichier étant toujours à la fin après un . , j'ai décidé de splité la liste d'obs. propre Resources avec .
#je suis ensuite allé compter le nombre de gif dans la liste SplitGif
    count=0
    for el in l:
        Split=el.split(".")
        if x in Split[-1]:
            count+=1
    return count

gif_count=file_count("gif",Resources)        
print(gif_count)
jpg_count=file_count("jpg",Resources)
print(jpg_count)
#Pour les observations propres nous avons donc 22 382 requetes d'accès à une image gif et 25 accès pour un jpg.

#Le plus complexe maintenant est de faire le décompte sur les observations "sales", heureusement il n'y en a que 129
#méthode bourrin, on va splitté tous les éléments de la liste Exceptions avec . afin d'être sur de faire ressortir le gif
jpg_count=file_count("jpg",Resources)

ExceptionsSplit=[]
for x in Exceptions:
    for y in x:
        SplitAll=y.split(".")
        ExceptionsSplit.extend(SplitAll)

gif_count2=ExceptionsSplit.count('gif"')+ExceptionsSplit.count("gif")
jpg_count2=ExceptionsSplit.count('jpg"')+ExceptionsSplit.count("jpg")
totalgif=gif_count+gif_count2
totaljpg=jpg_count+jpg_count2

print("Nous avons {} requêtes concernant les fichiers de type gif".format(totalgif))
print("Nous avons {} requêtes concernant les fichiers de type jpg".format(totaljpg))
print("Soit un total de {} requêtes concernant les fichiers de type images".format(totalgif+totaljpg))
```

```python
#INSTRUCTION 8

#Nous savons qu'il n'existe que 6 cas de code réponse 400
RC400_index=[]

for x,y in enumerate(ReplyCode):
    if y=="400":
        RC400_index.append(x)
print(RC400_index)


#Les codes 400 se trouvent dans les données mal splitées
#=>on a fusionné les ReplyCode des obs bien splitées de celles mal splitées dans l'instruction 2
#Les index 0 à 47618 correspondent donc aux obs propres
#Les index 47619 à 47747 correspondent aux obs sales

#On va maintenant aller chercher les indices des codes réponses 400 dans notre liste observations sales: Exceptions

Exceptions_RC_400_index=[]
for x,y in enumerate(RC400_index):
    exception_index=int(y)-47619
    Exceptions_RC_400_index.append(exception_index)
print(Exceptions_RC_400_index)    

#Creation d'une fonction permettant d'aller récupérer et stocker la taille des fichiers images de type x d'une liste l

def get_size(x,l):
    """fonction permettant de récupérer la taille d'un fichier de type x d'une liste l"""
    #Récupérons les index des requête concernant les images dans les obs propres
    index=[]
    for y,el in enumerate(l):
        Split=el.split(".")
        if x in Split[-1]:
            index.append(y)
    #Recupérons la tailles des images grâce aux index
    Filesize=[]      
    for el in index:
        Filesize.append(FileSize[el].replace("\n",""))
    return Filesize

Gif_FileSize=get_size("gif",Resources)
Jpg_FileSize=get_size("jpg",Resources)

print(Jpg_FileSize)


#Nous allons maintenant supprimé toutes les valeurs "-" et "0" de 
def clean_val(x,val):
    """"Supprime toutes les occurences val de la liste x"""
    while val in x:
            x.remove(val)
    return x


Gif_FileSize=clean_val(Gif_FileSize,"-")
Gif_FileSize=clean_val(Gif_FileSize,"0")
Jpg_FileSize=clean_val(Jpg_FileSize,"-")
Jpg_FileSize=clean_val(Jpg_FileSize,"0")


#Création d'une fonction qui permet de convertir les éléments d'une liste l en nombre entier

def convert_to_int(l):
    numeric=[]
    for el in l:
        numeric.append(int(el))
    return numeric

Gif_FileSize=convert_to_int(Gif_FileSize)
Jpg_FileSize=convert_to_int(Jpg_FileSize)

NumericSize=Gif_FileSize+Jpg_FileSize


#Désormais NumericSize contient toutes les tailles des images gif et jpg des observations propres

#Occupons nous maintenant de notre jeu de données sales: Exceptions
#Verifions tout d'abord que les indices 49 à 54 correspondent bien aux ReplyCode 400

for x in Exceptions_RC_400_index:
    print(Exceptions[x])

#On remarque que la taille renvoyée est "-", on peut donc garder ces élements dans notre liste. Il suffira de supprimer les 
# "-" plus tard

#Filtrons maintenant notre liste Exceptions afin qu'il ne reste plus que l'adresse de la page visitée

Filtered_list=[]
for x,y in enumerate(Exceptions):
    Filtered_list.append(y[3:-2])
    
#Cherchons maintenant les indices des pages contenant .gif et stockons les dans except_gif_index

except_gif_index=[]

for x,y in enumerate(Filtered_list):
    try:
        split=y[0].split(".")
    except IndexError:
        continue
    if "gif" in split[-1]:
        except_gif_index.append(x)

#Filtrons une derniere liste pour ne laisser que la taille des pages visitées contenant un gif

Exceptions_Filesize=[]
for x in except_gif_index:
    Exceptions_Filesize.append(Exceptions[x][-1].replace("\n",("")))
    
#Nous savons qu'il n'y a pas de .jpg dans la liste des Exceptions


#Nettoyons notre liste Exceptions_Filesize des valeurs "-" et "0"

Exceptions_Filesize=clean_val(Exceptions_Filesize,"-")
Exceptions_Filesize=clean_val(Exceptions_Filesize,"0")


#Nous allons maintenant convertir tous les element de Exceptions_Filesize du type str en type int

Exceptions_NumericSize=convert_to_int(Exceptions_Filesize)

#Output final

print("La taille moyenne des images demandées par les utilisateurs est de "+str(((sum(NumericSize)+sum(Exceptions_NumericSize))/(len(NumericSize)+len(Exceptions_NumericSize)))))
```
